**Рабочие формулы:**

**1)** $\omega_0 = 1 / \sqrt{L C} \Rightarrow L = \frac {1}{4 \pi f_{0n}^2 C}$ - собственная частота колебательного контура.

**2)** $\rho = \sqrt{L / C}$ - реактивное (волновое) сопротивление контура.

**3)** $R_\Sigma = R + R_L + R_S$ - суммарное активное сопротивление контура.

**4)** $Q = \rho / R_{\Sigma} = \omega_0 L / R_{\Sigma}$ - добротность. Можно посчитать добротность ещё как $Q = U_C(\omega_0) / E$, где $E$ - напряжение внешней ЭДС.

**5)** $R_{Smax} = 10 ^ {-3} \rho$ - максимальный вклад в активное сопротивление контура от конденсатора.

**6)** $R_L = R_\Sigma - R - R_{Smax}$ - оценочное активное сопротивление катушки.

**7)** $I = E / R_\Sigma$ - сила тока в контуре вблизи $\omega_0$.

```python
import numpy as np

# Calculate <L> and <R_L>
L_average = sum(L) / len(L)
R_L_average = sum(R_L) / len(R_L)

𝛼 = 0.95
L_std = 𝛼 * np.std(L)
R_L_std = 𝛼 * np.std(R_L)

sigma_C = [0.1 / i for i in C_n_list1] # From the photo.
sigma_omega_0 = [1 / i for i in f_0n_list1] # 1 Hz is the absolute error.
sigma_L = [(4 * sigma_omega_0[i]**2 + sigma_C[i]**2)**0.5 for i in range(len(f_0n_list1))]
sigma_R = [0.01 / i for i in R] # 0.01 Om is the absolute error
sigma_E = [0.03 for i in range(len(E_list1))] # From the documentation, 3%.
sigma_U_c = [0.03 for i in range(len(U_c_list1))] # From the documentation, 3%.
sigma_𝜌 = [(sigma_C[i]**2 + sigma_L[i]**2)**0.5 for i in range(len(𝜌))]
sigma_Q = [(sigma_E[i]**2 + sigma_U_c[i]**2)**0.5 for i in range(len(E_list1))]
sigma_R_sigma = [(sigma_𝜌[i]**2 + sigma_Q[i]**2)**0.5 for i in range(len(𝜌))]
sigma_R_L = [(sigma_𝜌[i]**2 + sigma_R_sigma[i]**2)**0.5 for i in range(len(𝜌))] # sigma_𝜌[i] == sigma_R_Smax[i]

sigma_L = max(sigma_L)
sigma_R_L = max(sigma_R_L)
print("Значение L: {:0.0f} ± {:0.0f} мкГн".format(L_average, (sigma_L**2 + L_std**2)**0.5))
print("Значение R_L: {:0.1f} ± {:0.1f} Ом".format(R_L_average, (sigma_R_L**2 + R_L_std**2)**0.5))
```
