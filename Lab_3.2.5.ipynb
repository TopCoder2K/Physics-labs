### –í—ã–Ω—É–∂–¥–µ–Ω–Ω—ã–µ –∫–æ–ª–µ–±–∞–Ω–∏—è –≤ —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–æ–º –ø–æ–ª–µ ###

**–¶–µ–ª—å:** –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –≤—ã–Ω—É–∂–¥–µ–Ω–Ω—ã—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π –∏ –ø—Ä–æ—Ü–µ—Å—Å–æ–≤ –∏—Ö —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏—è.

**–í —Ä–∞–±–æ—Ç–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è:** –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –ó–ß, –æ—Å—Ü–∏–ª–ª–æ–≥—Ä–∞—Ñ, –≤–æ–ª—å—Ç–º–µ—Ç—Ä, —á–∞—Å—Ç–æ—Ç–æ–º–µ—Ç—Ä, —ë–º–∫–æ—Å—Ç—å, –∏–Ω–¥—É–∫—Ç–∏–≤–Ω–æ—Å—Ç—å, –º–∞–≥–∞–∑–∏–Ω —Å–æ–ø—Ä–æ—Ç–∏–≤–ª–µ–Ω–∏–π, —É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω—ã–π –º–æ—Å—Ç.

–°—Å—ã–ª–∫–∞ –Ω–∞ –æ–ø–∏—Å–∞–Ω–∏–µ: https://mipt.ru/education/chair/physics/S_III/lab_el/new%203.2.5(4.6)_%D0%92%D1%8B%D0%BD%D1%83%D0%B6%D0%B4%D0%B5%D0%BD%D0%BD%D1%8B%D0%B5%20%D0%BA%D0%BE%D0%BB%D0%B5%D0%B1%D0%B0%D0%BD%D0%B8%D1%8F.pdf

**–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è:**

**1)** $\nu_0 = 1/(2\pi\sqrt{LC})$

**2)** $Q = \omega_0 / (2\Omega)$

**3)** $\theta = \frac{1}{n}\ln{\frac{U_0 - U_k}{U_0 - U_{k+n}}}$

**4)** $Q = \pi / \theta$

**5)** $Q = \frac{1}{R}\sqrt{\frac{L}{C}}$

**–•–æ–¥ —Ä–∞–±–æ—Ç—ã:**

**1)** –†–∞—Å—Å—á–∏—Ç–∞–µ–º —Ä–µ–∑–æ–Ω–∞–Ω—Å–Ω—É—é —á–∞—Å—Ç–æ—Ç—É –∫–æ–Ω—Ç—É—Ä–∞: $\nu_0 = 1/(2\pi\sqrt{LC}) = (1.59\pm0.03)\times10^3\:–ì—Ü$


–ú–µ–Ω—è—è —á–∞—Å—Ç–æ—Ç—É –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ –≤ –æ–±–µ —Å—Ç–æ—Ä–æ–Ω—ã –æ—Ç —Ä–µ–∑–æ–Ω–∞–Ω—Å–Ω–æ–π –∫—Ä–∏–≤–æ–π, —Å–Ω–∏–º–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø–æ–∫–∞–∑–∞–Ω–∏–π –≤–æ–ª—å—Ç–º–µ—Ç—Ä–∞ $U$ –æ—Ç –ø–æ–∫–∞–∑–∞–Ω–∏–π —á–∞—Å—Ç–æ—Ç–º–µ—Ç—Ä–∞ $ \nu $ –¥–ª—è $R = 0\:–û–º$ –∏ $R = 100\:–û–º$

```python
from math import pi
L_1 = 0.1 # –ì–Ω
C_1 = 0.1 / 1000000 # –º–∫–§ —Ç–æ—á–Ω–æ—Å—Ç—å 2 %
nu_0_1 = 1 / (2 * pi * (L_1 * C_1)**(1/2))
print(nu_0_1)

print(22 * 0.3)
print(22 * 0.7 / 7 / 2)

# —Ç–æ—á–Ω–æ—Å—Ç—å –†567: (2 + 0.08/L) %
sigma_L_1 = 2.0008 / 100 * L_1 # –º–ì–Ω
sigma_C_1 = 0.02*0.1/1000000
sigma_nu_0_1 = (sigma_L_1**2/(16*pi**2*C_1*L_1**3) + sigma_C_1**2/(16*pi**2*L_1*C_1**3))**(1/2)
print(sigma_nu_0_1)
# —É—Å—Ç–∞–Ω–æ–≤–∏–ª–∏ nu_0_1 = 1597 –ì—Ü —Ç–æ—á–Ω–æ—Å—Ç—å 1
```

```python
import pandas as pd

# R = 0 
nu_list1_1 = pd.Series([ 1597, 1609, 1617, 1625, 1635, 1649, 1675, 1708, 1583, 1575, 1567, 1558, 1548, 1528, 1509 ])
#U_list1_1 = [ 29, 25.5, 23, 19.5, 16.5, 13, 9.5, 7, 25.5, 23, 19.5, 16.5, 13.5, 10.5, 7 ]
# –º–Ω–æ–∂–∏—Ç–µ–ª—å –æ—Å—Ü–∏–ª–ª–æ–≥—Ä–∞—Ñ–∞: 0.5 V
# –º–Ω–æ–∂–∏—Ç–µ–ª—å –≤–æ–ª—å—Ç–º–µ—Ç—Ä–∞: 1 V
U_list1_1 = pd.Series([ 31, 28, 25, 22, 18.5, 15, 11, 8, 28.5, 25.5, 22, 18.5, 15.5, 11.5, 9 ]) # –í

#R = 100

# –º–Ω–æ–∂–∏—Ç–µ–ª—å –æ—Å—Ü–∏–ª–ª–æ–≥—Ä–∞—Ñ–∞: 0.2 V
# –º–Ω–æ–∂–∏—Ç–µ–ª—å –≤–æ–ª—å—Ç–º–µ—Ç—Ä–∞: 0.3 V
nu_list1_2 = pd.Series([ 1597, 1651, 1686, 1711, 1752, 1807, 1878, 1966, 1553, 1531, 1509, 1483, 1451, 1418, 1380 ])
U_list1_2 = pd.Series([ 31, 28.5, 25, 22.5, 19.5, 16.5, 13.5, 11.5, 28, 25.5, 22.5, 19.5, 16.5, 13.5, 11 ])

table_1 = pd.DataFrame({
    "ùúà / ùúà0" : round(nu_list1_1 / nu_0_1, 4),
    "U / U_0" : round(U_list1_1 / U_list1_1[0], 4)
})

print("R = 0 –û–º")
print(table_1.T)

table_2 = pd.DataFrame({
    "ùúà / ùúà0" : round(nu_list1_2 / nu_0_1, 4),
    "U / U_0" : round(U_list1_2 / U_list1_2[0], 4)
})

print("\nR = 100 –û–º")
print(table_2.T)
```

```python
val_x = table_1["ùúà / ùúà0"]
val_y = table_1["U / U_0"]

val_x_1 = table_2["ùúà / ùúà0"]
val_y_1 = table_2["U / U_0"]

# –°—á–∏—Ç–∞–µ–º –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç–∏ –∏–∑–º–µ—Ä–µ–Ω–∏—è U/U_0 –∏ nu/nu_0
# sigma_nu_list1_1 = ( 1 / (len(nu_list1_1) - 1) * sum((nu_list1_1 - mean(nu_list1_1))**2) )**(1/2)
sigma_val_x = val_x * ( (1 / nu_list1_1)**2 + (1 / nu_list1_1[0])**2 )**(1/2);
sigma_val_y = val_y * ( (1 / U_list1_1)**2 + (1 / U_list1_1[0])**2 )**(1/2);

sigma_val_x_1 = val_x_1 * ( (1 / nu_list1_2)**2 + (1 / nu_list1_2[0])**2 )**(1/2);
sigma_val_y_1 = val_y_1 * ( (1 / U_list1_2)**2 + (1 / U_list1_2[0])**2 )**(1/2);

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 16 # –£–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–∏–ª–µ–º, –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ–º - —Ä–∞–∑–º–µ—Ä–æ–º —à—Ä–∏—Ñ—Ç–∞ 
 # –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É
plt.figure(figsize=(10,7))

# –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º –æ—Å–∏ –∏ –≥—Ä–∞—Ñ–∏–∫
plt.title(r"–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤-—Ç–∏ –æ—Ç–Ω–æ—à–µ–Ω–∏—è –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–π –∫ —Ä–µ–∑–æ–Ω–∞–Ω—Å–Ω–æ–º—É $U / U_0$ –æ—Ç –æ—Ç–Ω–æ—à–µ–Ω–∏—è —á–∞—Å—Ç–æ—Ç –∫ —Ä–µ–∑–æ–Ω–∞–Ω—Å–Ω–æ–π $\nu / \nu_0$ ")
plt.ylabel("$U/U_0$")
plt.xlabel(r"$\nu/\nu_0$")

# –î–∞–Ω–Ω—ã–µ
x1 = val_x
y1 = val_y
plt.plot(x1, y1, 'r^', label='R = 0 –û–º')

x2 = val_x_1
y2 = val_y_1
plt.plot(x2, y2, 'bs', label='R = 100 –û–º')
# 'r^' - –∑–∞–¥–∞–µ—Ç —Å—Ç–∏–ª—å –ª–∏–Ω–∏–∏ - –∫—Ä–∞—Å–Ω—ã–µ (red) —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∏ (^), –ø–æ–¥—Ä–æ–±–Ω–µ–µ –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏

# –î–∞–Ω–Ω—ã–µ —Å –æ—à–∏–±–∫–∞–º–∏
# mu = np.sin(x2)
# sigma = np.abs(mu)**0.5
# y2 = np.random.normal(mu, sigma)
# –ú–æ–∂–Ω–æ —Ä–∏—Å–æ–≤–∞—Ç—å –æ—à–∏–±–∫–∏
plt.errorbar(x1, y1, yerr=sigma_val_y, xerr=sigma_val_x, fmt='.', label='–ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å') 
plt.errorbar(x2, y2, yerr=sigma_val_y_1, xerr=sigma_val_x_1, fmt='.', label='–ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å') 

# –ê–∫—Ç–∏–≤–∏—Ä—É–µ–º —Å–µ—Ç–∫—É
plt.grid(b=True, which='major', axis='both', alpha=1)
plt.grid(b=True, which='minor', axis='both', alpha=0.5)

# –ê–∫—Ç–∏–≤–∏—Ä—É–µ–º –ª–µ–≥–µ–Ω–¥—É –≥—Ä–∞—Ñ–∏–∫–∞
plt.legend()
# –í–Ω–∏–º–∞–Ω–∏–µ, –∑–∞–ø—É—Å–∫–∞–µ—Ç–µ –≤–∞—à—É –ø—Ä–æ–≥—Ä–∞–º–º—É –∫–∞–∫ —Å—Ü–µ–Ω–∞—Ä–∏–π, —Ç–æ —á—Ç–æ –±—ã –ø–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —ç—Ç—É –∫–æ–º–∞–Ω–¥—É
## plt.show()
```

–†–∞—Å—Å—á–∏—Ç–∞–µ–º –¥–æ–±—Ä–æ—Ç–Ω–æ—Å—Ç—å –∫–æ–Ω—Ç—É—Ä–∞ –ø—Ä–∏: 
    
$R = 0\:–û–º$ –∏–º–µ–µ–º $Q = (2.67\pm0.04)\times10^1$
    
$R = 100\:–û–º$ –∏–º–µ–µ–º $Q = 7\pm2$
    
–°—Ä–∞–≤–Ω–∏–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –∏ —ç–∫—Å–ø—Ä–µ–∏–º–µ–Ω—Ç–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑–æ–Ω–∞–Ω—Å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã: —Å —É—á—ë—Ç–æ–º –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω–æ–µ –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–µ–∑–æ–Ω–∞–Ω—Å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã —Å–æ–≤–ø–∞–¥–∞—é—Ç

```python
# –†–∞—Å—á—ë—Ç –¥–æ–±—Ä–æ—Ç–Ω–æ—Å—Ç–∏ Q
# –ó–¥–µ—Å—å –ø—Ä–∏–Ω—Ü–∏–ø –Ω—É–º–µ—Ä–∞—Ü–∏–∏ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –æ—Å—Ç–∞–ª—å–Ω–æ–π —á–∞—Å—Ç–∏ —Ä–∞–±–æ—Ç—ã (–¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞)!!! 
# –°—á–∏—Ç–∞–µ–º Omega (x_3 –∏ x_3_1), –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º U/U0 = 0.7
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –ø–æ–ª–æ–∂–µ–Ω–∏–µ —ç—Ç–∏—Ö —Ç–æ—á–µ–∫ (x_1, y_1) –∏ (x_2, y_2), –ø—Ä–æ–≤–µ–¥—è –ø—Ä—è–º—É—é –ø–æ –¥–≤—É–º –±–ª–∏–∂–∞–π—à–∏–º —Ç–æ—á–∫–∞–º
# –∫ –∏—Å–∫–æ–º–æ–π (x_3, 0.7)

# R = 0 –û–º

# –õ–µ–≤–∞—è —Ç–æ—á–∫–∞
x_1 = 2 * pi * nu_list1_1[10]
x_2 = 2 * pi * nu_list1_1[11]
y_1 = table_1["U / U_0"][10]
y_2 = table_1["U / U_0"][11]
x_3 = x_1 + (x_2 - x_1) / (y_2 - y_1) * (0.7 - y_1)

s_x = 1 # sigma_x, –ì—Ü
s_y = pd.Series([sigma_val_x[10], sigma_val_x[11]]).max() # sigma_y
sigma_x_3 = ( ( 2*(s_x/(x_2-x_1))**2 + 2*(s_y/(y_2-y_1))**2 + 2*(s_y/(0.7-y_1))**2 )*(x_3 - x_1)**2 + s_x**2 )**(1/2)

# –ü—Ä–∞–≤–∞—è —Ç–æ—á–∫–∞ (–≤—Å–µ –ø–µ–µ–º–µ–Ω–Ω—ã–µ —Å —Å—É—Ñ—Ñ–∏–∫—Å–æ–º "_1")
x_1_1 = 2 * pi * nu_list1_1[3]
x_2_1 = 2 * pi * nu_list1_1[4]
y_1_1 = table_1["U / U_0"][3]
y_2_1 = table_1["U / U_0"][4]
x_3_1 = x_1_1 + (x_2_1 - x_1_1) / (y_2_1 - y_1_1) * (0.7 - y_1_1)

s_x_1 = 1 # sigma_x, –ì—Ü
s_y_1 = pd.Series([sigma_val_x_1[10], sigma_val_x_1[11]]).max() # sigma_y
sigma_x_3_1 = ( ( 2*(s_x_1/(x_2_1-x_1_1))**2 + 2*(s_y_1/(y_2_1-y_1_1))**2 + 2*(s_y_1/(0.7-y_1_1))**2 )*(x_3_1 - x_1_1)**2 + s_x_1**2 )**(1/2)

Q_1 = 2 * pi * nu_0_1 / (x_3_1 - x_3)
sigma_Q_1 = Q_1 * ( (sigma_nu_0_1/nu_0_1)**2 + ((sigma_x_3)**2 + (sigma_x_3_1)**2)/(x_3_1-x_3)**2 )**(1/2)
print(Q_1)
print(sigma_Q_1)

# R = 100 –û–º

# –õ–µ–≤–∞—è —Ç–æ—á–∫–∞
x_1 = 2 * pi * nu_list1_2[10]
x_2 = 2 * pi * nu_list1_2[11]
y_1 = table_2["U / U_0"][10]
y_2 = table_2["U / U_0"][11]
x_3 = x_1 + (x_2 - x_1) / (y_2 - y_1) * (0.7 - y_1)

s_x = 1 # sigma_x, –ì—Ü
s_y = pd.Series([sigma_val_x[10], sigma_val_x[11]]).max() # sigma_y
sigma_x_3 = x_3 * ( 2*(s_x/(x_2-x_1))**2 + 2*(s_y/(y_2-y_1))**2 + 2*(s_y/(0.7-y_1))**2 + (s_x/x_1)**2 )**(1/2)

# –ü—Ä–∞–≤–∞—è —Ç–æ—á–∫–∞ (–≤—Å–µ –ø–µ–µ–º–µ–Ω–Ω—ã–µ —Å —Å—É—Ñ—Ñ–∏–∫—Å–æ–º "_1")
x_1_1 = 2 * pi * nu_list1_2[3]
x_2_1 = 2 * pi * nu_list1_2[4]
y_1_1 = table_2["U / U_0"][3]
y_2_1 = table_2["U / U_0"][4]
x_3_1 = x_1_1 + (x_2_1 - x_1_1) / (y_2_1 - y_1_1) * (0.7 - y_1_1)

s_x_1 = 1 # sigma_x, –ì—Ü
s_y_1 = pd.Series([sigma_val_x_1[10], sigma_val_x_1[11]]).max() # sigma_y
sigma_x_3_1 = x_3_1 * ( 2*(s_x_1/(x_2_1-x_1_1))**2 + 2*(s_y_1/(y_2_1-y_1_1))**2 + 2*(s_y_1/(0.7-y_1_1))**2 + (s_x_1/x_1_1)**2 )**(1/2)

Q_2 = 2 * pi * nu_0_1 / (x_3_1 - x_3)
sigma_Q_2 = Q_1 * ( (sigma_nu_0_1/nu_0_1)**2 + (sigma_x_3/x_3)**2 + (sigma_x_3_1/x_3_1)**2 )**(1/2)
print(Q_2)
print(sigma_Q_2)
```
