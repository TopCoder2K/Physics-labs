### "Сдвиг фаз в цепи переменного тока"

**Цель работы: изучить влияние активного сопротивления, индуктивности и ёмкости на сдвиг фаз между током и напряжением в цепи переменного тока**

**В работе используются: генератор звуковой частоты (ЗГ), двухканальный осциллограф (ЭО), магазин ёмкостей, магазин сопротивлений, катушка индуктивности, резисторы, мост переменного тока.**

**Рабочие формулы**: 1) $\alpha = \pm \arcsin{\frac{y_{x = 0}}{y_0}}$, где $\alpha$ - сдвиг фаз между напряжениями, где $y_0$ - максимальное отклонение по оси $y$. 2) $2 \arctan{\frac{1}{\Omega R C}}$ - сдвиг фаз между входным и выходным напряжениями (для фазовращателя), где $\Omega$ - частота внешнего источника, $R$ - переменное сопротивление, $C$ - ёмкость конденсатора. 3) $\psi = \arctan{\frac{\Omega L - 1\ /\ \Omega C}{R}}$ - отставание тока от напряжения по фазе в контуре из последовательно соединённых элементов.

**Схема установки и описание работы**: https://mipt.ru/education/chair/physics/S_III/lab_el/new%203.2.1%20(4.7)_%D0%A1%D0%B4%D0%B2%D0%B8%D0%B3%20%D1%84%D0%B0%D0%B7%20%D0%B2%20%D1%86%D0%B5%D0%BF%D0%B8%20%D0%BF%D0%B5%D1%80%D0%B5%D0%BC%D0%B5%D0%BD%D0%BD%D0%BE%D0%B3%D0%BE%20%D1%82%D0%BE%D0%BA%D0%B0.pdf

**1) Исследование зависимости сдвига фаз между током и напряжением  от $R$ в $RC$-цепи**

1.1) Рассчитаем реактивное сопротивление цепи: $X_1 = 1 \ / \ (\Omega \cdot C)$, где $\Omega = 2 \cdot \pi \cdot \nu$ - циклическая частота: $X_1 = 318 \pm 1\ Ом$

1.2) Увеличивая сопротивление $R$ от нуля до $10 \cdot X_1$, проведём измерение сдвига фаз. Заметим, что одно деление - $0.1\ ms$

```python
import pandas as pd

pi = 3.14159265359

resistances = pd.Series([0, 530, 1060, 1590, 2120, 2650, 3180])
x0 = pd.Series([24, 24, 23, 23, 23, 23, 23])
x = pd.Series([12, 4, 3, 2, 2, 1, 1])

## Зависимость сдвига фаз между током и напряжением от  𝑅  в  𝑅𝐶 -цепи.
table1 = pd.DataFrame({
    'Resistances, Om': resistances,
    'x0, divisions' : x0,
    'x, divisions' : x,
    'psi' : x / x0 * pi
})

print(table1)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import interpolate

mpl.rcParams['font.size'] = 16 # Change size font.
# Create figure.
plt.figure(figsize=(15, 15))

# Sign axises and the graph.
plt.title("График 1, $\psi\ =\ f(R_\Sigma)$ в $RC$-цепи")
plt.ylabel("$\psi$, рад", fontsize = 20)
plt.xlabel("$R_\Sigma$, Ом", fontsize = 20)

# Get summary resistance.
r = 12.2

x_resist = table1['Resistances, Om'] + r
y_psi = table1['psi']
plt.scatter(x_resist, y_psi)
# plt.plot(x_resist, y_psi, 'o')

# TRYYYYYYYYYYYYYIN TO  the least-squares solution.
# a, b = np.linalg.lstsq(x_resist, y_psi, rcond=None)
# Interpolation by quadratic function.
f = interpolate.interp1d(x_resist, y_psi, kind="quadratic")
x_new = np.linspace(x_resist.min(), x_resist.max(), 100)
plt.plot(x_new, f(x_new))

# Calculate errors. As resistance is relatively accurate, psi is take into account.
sigma_x_abs = pd.Series([0.5 for i in range(len(table1['psi']))])
sigma_x_rel = sigma_x_abs / table1['x, divisions']

sigma_y = y_psi * sigma_x_rel

# Draw errors.
plt.errorbar(x_resist, y_psi, yerr=sigma_y, fmt='.', label='Погрешности') 

# Activate grid.
plt.grid(b=True, which='major', axis='both', alpha=1)

# Activate legend.
plt.legend()
```
